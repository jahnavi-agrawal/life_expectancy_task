# Explaining the data set:

Country: Country under study

Year: year

Status: Status of the country's development (Developing/Developed)

Population: Population of country

Hepatitis B: Percentage of 1-year-olds immunized against hepatitis B

Measles: The number of reported measles cases per 1000 people

Polio: Percentage of 1-year-olds immunized against polio

Diphtheria: Percentage of 1-year-olds immunized against diphtheria

HIV/AIDS: The number of deaths caused by AIDS of the last 4-year-olds who were born alive per 1000 people

infant deaths: The number of infant deaths per 1000 people

under-five deaths: The number of deaths of people under 5 years old per 1000 people

Total expenditure: The ratio of government medical-health expenses to total government expenses in percentage

GDP: Gross domestic product

BMI: The average body mass index of the entire population of the country

thinness 1-19 years: Prevalence of thinness among people 19 years old in percentage

Alcohol: Liters of alcohol consumption among people over 15 years old

Schooling: The number of years that people study

Life expectancy: Country life expectancy

**Handling Missing Values:**

```python
missing = data.isnull().sum()
missing_p = (missing / len(data)) * 100

missing_data = pd.DataFrame({
    'Missing Values': missing,
    'Missing Percentage': missing_p
})

missing_data = missing_data.sort_values(by='Missing Percentage', ascending=False)
print(missing_data)
```

None of the features have very high missing values, so we can proceed with our data, without dropping any column.


**Handling missing values:**

```python
# num= numerical columns
num = data.select_dtypes(include=[np.number]).columns
for col in num:
   data[col] = data[col].fillna(data[col].mean())


# cat= categorical columns 
cat = data.select_dtypes(include=['object']).columns
for col in cat:
    data[col] = data[col].fillna(data[col].mode()[0])

```

**Visualising data:**

```python
cols = ["Adult Mortality", "Schooling", "HIV/AIDS", "Income composition of resources", "Total expenditure", "Alcohol"]
plt.figure(figsize=(15, 20))
for i, col in enumerate(cols, 1):
    plt.subplot(5, 3, i)
    plt.hist(data[col])
    plt.title(col)
```

**Metrics:**

It is obvious, a developed country would have proper resources and awareness that can ensure a better life expectancy. Also the trend chart below will show how average life expectancy has improved over the years.

```python
avg_by_year = data.groupby("Year")["Life expectancy"].mean()
sns.lineplot(x=avg_by_year.index, y=avg_by_year.values)
plt.title("Average Life Expectancy Over Time")
plt.show()
```

```python
corr = data.select_dtypes(include=['number']).corr()
life_corr = corr["Life expectancy"].drop("Life expectancy")
life_corr_sorted = life_corr.sort_values(ascending=False). round(3)

print("Correlation of features with Life Expectancy:")
print(life_corr_sorted)
```

On paying attention to the "Life expectancy" row, we find that there is negative correlation with constructs like Adult Mortality, HIV, malnutrition etc and positive correlation with Schooling, Income composition, BMI etc. 

```python
plt.figure(figsize=(10,6))
life_corr_sorted.plot(kind='bar', color=['skyblue' if v > 0 else 'salmon' for v in life_corr_sorted])
plt.title("Correlation of Features with Life Expectancy", fontsize=14)
plt.ylabel("Correlation Coefficient", fontsize=12)
plt.xticks(rotation=75, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
cont_vars = list(data.columns)[3:]
mask = np.triu(data[cont_vars].corr())
plt.figure(figsize=(15,6))
sns.heatmap(data[cont_vars].corr(), annot=True, fmt='.2g', vmin=-1, vmax=1, center=0, cmap='coolwarm', mask=mask)
plt.ylim(18, 0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

# Insights:


a) Education (Schooling, 0.76) and income equality (0.73) are the strongest positive drivers of life expectancy.
