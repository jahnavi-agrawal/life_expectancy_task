# **Exploration Data Analysis:**
**Import packages:**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="whitegrid")
```

**Load Dataset:**

```python
data_path = "E:\Visual Studio Code\IML\Jahnavi_Neha_A1\life_expectancy_task\data\Life Expectancy Raw.csv"
data = pd.read_csv(data_path)
print("Loaded Raw Dataset:", data_path)
```

**Handling Missing Values:**

```python
data.isnull().sum().sort_values(ascending=False)
```

**Dropping Target variable:**

```python
target= data["Life expectancy"]
#data= data.drop(columns=["Country", "Year", "Measles"])
#removed measles because it was an absolute value
data.shape
```

**Filling Missing Value:**

```python
# Replacing missing values with mean
# num= numerical columns
num = data.select_dtypes(include=[np.number]).columns
for col in num:
    data[col].fillna(data[col].mean())
```

```python
# Filling categorical features with mode
# cat= categorical columns 
cat = data.select_dtypes(include=['object']).columns
for col in cat:
    data[col].fillna(data[col].mode()[0])
```

**Verifying Missing Values**

```python
print(data.isnull().sum())
```

**Metrics:**

```python
print(data.groupby("Status")["Life expectancy"].mean())
```

It is obvious, a developed country would have proper resources and awareness that can ensure a better life expectancy. Also the trend chart below will show how average life expectancy has improved over the years.

```python
avg_by_year = data.groupby("Year")["Life expectancy"].mean()
sns.lineplot(x=avg_by_year.index, y=avg_by_year.values)
plt.title("Average Life Expectancy Over Time")
plt.show()
```

```python
corr = data.select_dtypes(include=['number']).corr()
life_corr = corr["Life expectancy"].drop("Life expectancy")
life_corr_sorted = life_corr.sort_values(ascending=False). round(3)

print("Correlation of features with Life Expectancy:")
print(life_corr_sorted)
```

On paying attention to the "Life expectancy" row, we find that there is negative correlation with constructs like Adult Mortality, HIV, malnutrition etc and positive correlation with Schooling, Income composition, BMI etc. 

```python
plt.figure(figsize=(10,6))
life_corr_sorted.plot(kind='bar', color=['skyblue' if v > 0 else 'salmon' for v in life_corr_sorted])
plt.title("Correlation of Features with Life Expectancy", fontsize=14)
plt.ylabel("Correlation Coefficient", fontsize=12)
plt.xticks(rotation=75, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
cols = ["Adult Mortality", "Schooling", "HIV/AIDS", "Income composition of resources", "Total expenditure", "Alcohol"]
fig, axes = plt.subplots(3, 2, figsize=(12, 10))

for i, col in enumerate(cols):
    row, col_pos = divmod(i, 2)  # find row and column index
    sns.scatterplot(x=data[col], y=data["Life expectancy"], ax=axes[row, col_pos], color= "teal")
    axes[row, col_pos].set_title(f"Life Expectancy vs {col}")

plt.tight_layout()
plt.show()
```

```python
selected=["Adult Mortality", "Schooling", "HIV/AIDS", "Income composition of resources", "BMI"]

data[selected].hist(figsize=(12,8), bins=30, edgecolor="black")
plt.tight_layout()
plt.show()
```

# Insights:


a) Education (Schooling, 0.76) and income equality (0.73) are the strongest positive drivers of life expectancy.

b) Higher vaccination rates (Diphtheria, Polio, Hepatitis B) and healthcare spending also improve life expectancy.

c) Adult Mortality (-0.69), HIV/AIDS (-0.56), and child malnutrition (thinness, -0.47) are the strongest negative factors.

d) GDP matters (0.46), but social and health factors (education, disease prevention) play a larger role.
